```python
df_unique_solvedPerMonth_Rate = df.drop_duplicates(subset=["Number_of_solved_problems", "Max_Rate"])

plt.scatter(df_unique_solvedPerMonth_Rate["Number_of_solved_problems"], df_unique_solvedPerMonth_Rate["Max_Rate"],color="palevioletred")
plt.xlabel("Number of solved problems")
plt.ylabel("Max Rate")
plt.title("Relationship Between Solved Problems and Maximum Codeforces Rating")

```

```python
df['University'] = df['University'].str.strip()
df["University"].unique()
```

```python
selected_universities = ["Aleppo University", "Egypt Universities","Homs University","Latakia University","Damascus University","Secondary school"]
selected_rates = ["Newbie","Pupil", "Specialist", "Expert"]

# Filter
df_filtered = df[
    df["University"].isin(selected_universities) &
    df["Rate"].isin(selected_rates)
]

# Count unique students
counts = df_filtered.groupby(["University", "Rate"])["Student_ID"].nunique().reset_index()

plt.figure(figsize=(8, 6))

sns.barplot(
    data=counts,
    x="University",
    y="Student_ID",
    hue="Rate",
    palette="pastel"
)

plt.title("Number of Girls by University and Rate")
plt.xlabel("University")
plt.ylabel("Number of Girls")
plt.legend(title="Rate")

plt.xticks(rotation=45, ha='right')  # تدوير النص
plt.tight_layout()  # ضبط المسافات
```

```python
plt.figure(figsize=(6, 4))
corr = np.corrcoef(df['Max_Rate'], df['Minutes_spent_trying_before_seeking_help'])

sns.heatmap(corr, annot=True, cmap="coolwarm",
            xticklabels=["Max_Rate","Minutes_spent_trying_before_seeking_help"],
            yticklabels=["Max_Rate","Minutes_spent_trying_before_seeking_help"])

plt.title("Correlation Heatmap Between Two Fields")
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()  
```

```python
avg_rate = df.groupby(["Previous_competitions"])["Max_Rate"].mean().reset_index()

plt.figure(figsize=(6, 6))
sns.barplot(
    data=avg_rate,
    x="Previous_competitions",
    y="Max_Rate",
    color="palevioletred"
)
plt.xlabel("Previous Competition")
plt.ylabel("Average Max Rate")
plt.title("Average Max Rate by Previous Competition")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

```python
df['Reason_for_not_benefiting_from_training_camps'] = df['Reason_for_not_benefiting_from_training_camps'].str.strip()

plt.figure(figsize=(8, 6))
Reason_for_not_benefiting_from_training_camps = df.groupby('Reason_for_not_benefiting_from_training_camps')["Student_ID"].nunique()
plt.bar(Reason_for_not_benefiting_from_training_camps.index,
        Reason_for_not_benefiting_from_training_camps.values, color='palevioletred')

plt.title('Reasons for Not Benefiting from Training Camps')

plt.xticks(rotation=45, ha='right')  # تدوير النص
plt.tight_layout()  # ضبط المسافات
```

```python
df['Reasons_hindering_training'] = df['Reasons_hindering_training'].str.strip()

plt.figure(figsize=(8, 8))
Reasons_hindering_training = df.groupby('Reasons_hindering_training')["Student_ID"].nunique()
plt.bar(Reasons_hindering_training.index,
        Reasons_hindering_training.values, color='palevioletred')

plt.title('Main Factors Hindering Training')

plt.xticks(rotation=45, ha='right')  # تدوير النص
plt.tight_layout()  # ضبط المسافات
```
